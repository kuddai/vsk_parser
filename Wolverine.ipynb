```python
%pylab inline

from matplotlib import pyplot as plt
from matplotlib import animation
#plt.rcParams['animation.ffmpeg_path'] = 'C:\\ffmpeg\\bin\\'

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim, fps=120):
    if not hasattr(anim, '_encoded_video'):
        anim.save("test.mp4", fps=fps, extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])

        video = open("test.mp4","rb").read()

    anim._encoded_video = video.encode("base64")
    return VIDEO_TAG.format(anim._encoded_video)

from IPython.display import HTML

def display_animation(anim, fps=120):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim, fps=fps))
```

```python
from vicon_anim_parser.vsk_parser import parse_child2parent
from vicon_anim_parser.csv_anim_parser import parse_segments_names, remove_subjects_names

header = parse_segments_names("./SwordMovements/sword_movements_01.csv")
header = remove_subjects_names(header)
#remove sword segment
header = header[:-1]

child2parent = parse_child2parent("Ruslan.vsk")

segment_name2segment_id = {segment_name: segment_id for segment_id, segment_name in enumerate(header)}

segment_id2parent_id = [None] * len(header)
for segment_id in xrange(len(header)):
    segment_name = header[segment_id]
    parent_name = child2parent[segment_name]
    parent_id = -1 if parent_name is None else segment_name2segment_id[parent_name]
    segment_id2parent_id[segment_id] = parent_id

print segment_id2parent_id
```
