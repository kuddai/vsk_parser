```python
%matplotlib inline
#common plotting rutines dfdf bbh

import sys
import matplotlib.pyplot as plt

def plot_horizontal_data(mouse_data, line_color="r", marker_color="g", linewidth=4.0):
    xx = [r[0] for r in mouse_data]
    vx = [r[2] for r in mouse_data]
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    ax.grid(True, which='both')
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.plot(xx, vx, "-", c=line_color, linewidth=linewidth)
    ax.plot(xx, vx, "o", c=marker_color)
    ax.set_xlabel("Position X")
    ax.set_ylabel("Velocity X")
    
    plt.show()

def parse_mouse_data(mouseDataFileName):
    mouseData = []
    with open(mouseDataFileName) as f:
        #skip header
        f.readline()
        for line in f:
            try:
                mouseData.append(map(float, line.split(",")))
            except:
                print line
                raise ValueError("problem on " + line)
    return mouseData
```

```python
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 8, 6  
```

```python
plot_horizontal_data(parse_mouse_data("./MouseCurvesRaw/Curve_PD_Friction_Left2Right_Slow.txt"))
```
